4. Wrangle the data so that it follows consistent logic; for example, rename columns with illogical names and drop columns that don’t add anything to your analysis.

:: First Name = first_name, Surnam = last_name, Gender = gender, STATE = state, Age = age, n_dependants = num_dep, fam_status = marital_status

```python
# Rename columns

# First Name = first_name
df_cust.rename(columns = {'First Name' : 'first_name'}, inplace = True)

# Surnam = last_name
df_cust.rename(columns = {'Surnam' : 'last_name'}, inplace = True)

# Gender = gender
df_cust.rename(columns = {'Gender' : 'gender'}, inplace = True)

# STATE = state
df_cust.rename(columns = {'STATE' : 'state'}, inplace = True)

# Age = age
df_cust.rename(columns = {'Age' : 'age'}, inplace = True)

# n_dependants = num_dep
df_cust.rename(columns = {'n_dependants' : 'num_dep'}, inplace = True)

# fam_status = marital 
df_cust.rename(columns = {'fam_status' : 'marital_status'}, inplace = True)
```

5. Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
# Check for missing values

df_cust.isnull().sum()
```

:: The column 'first_name' contains missing values. Opt to leave as the customer first name is not as relevant to analysis as other column data. However, will create df_cust_nan for further analysis and future use.

```python
# Create df for missing values in column 'first_name'

df_cust_nan = df_cust[df_cust['first_name'].isnull() == True]
```

```python
# Check df_cust_nan

df_cust_nan.shape
```

```python
# Check for mixed data

for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

:: The column 'first_name' contains mixed data. 

```python
# Change mixed type column 'first_name' to str

df_cust['first_name'] = df_cust['first_name'].astype('str')
```

:: The column 'user_id' should be changed to str since the values are spcific to the user and not used for analysis.

```python
# Change column 'user_id' type to str

df_cust['user_id'] = df_cust['user_id'].astype('str')
```

```python
# Check column type changes

df_cust.dtypes
```

```python
# Check for duplicates

df_dups = df_cust[df_cust.duplicated()]
```

```python
df_dups.head()
```

:: No apparent duplicates

6. Combine your customer data with the rest of your prepared Instacart data.

```python
# Import 'ords_prod_merg48' for combine.

ords_prods_merg48 = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merg48.pkl'))
```

```python
# Check import

ords_prods_merg48.head()
```

```python
# Check types for combine

ords_prods_merg48.dtypes
```

```python
# Change column 'user_id' type to str

ords_prods_merg48['user_id'] = ords_prods_merg48['user_id'].astype('str')
```

```python
# Drop existing '_merge' column

ords_prods_merg48.drop(columns = ['_merge'])
```

```python
# Merge new 'customer' df with existing 'ords_prods_merg48' using left join to add 'df_cust' data to full 'ords_prods_merge48' data

df_insta_full = ords_prods_merg48.merge(df_cust, on = 'user_id', how = 'inner')
```
