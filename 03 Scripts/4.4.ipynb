Resetting index. 
Header on new transposed df shows '0', need to remove and index the rows.

:: First index the rows

```python
# This will index the rows of the transposed data, using .reset_index()
df_dep_t.reset_index()
```

:: Step 1, new header

```python
# .iloc[] = LOCate an Integer. Using this function, you can locate the row to assign a new header. 
# Note, brakets not parenthesis
new_header = df_dep_t.iloc[0]
```

```python
new_header
```

:: Step 2, new df. Copy the existing df from row 1 on to end

```python
# the function [1:] tells python to copy from row 1 to end
df_dep_t_new = df_dep_t[1:]
```

```python
df_dep_t_new
```

:: Step 3, insert new_header

```python
# using .columns you can insert new names to columns. 
# In this line, by using 'new_header' that you created, columns are renamed to the new header
df_dep_t_new.columns = new_header
```

```python
df_dep_t_new
```

Note: the dataframe df_dep_t_new is missing the index column here. This could be added by .reset_index() .

Subsetting

:: Indexing - to retrieve an element from an object. syntax: object['element']

```python
df_prods['department_id']==19
```

:: Nesting syntax above inside df[ ] will return rows where only the [dept_id] == 19 is True

```python
df_prods[df_prods['department_id']==19]
```

:: Once the filter is defined you can create the subset df - df_snacks = df[df['element']==identifier]

```python
df_snacks = df_prods[df_prods['department_id']==19]
```

:: Dataframe = all items in the snack department!

```python
df_snacks.head()
```

:: Same subsetting using loc function. loc will look for the column. 

```python
# Creating same df with different syntax and name for practice
df_snacks_2 = df_prods.loc[df_prods['department_id']==19]
```

```python
df_snacks_2.head()
```

:: Another alt for subsetting a df - using .isin can pull multiple identifiers

:: Ex - .isin[(17,18,19)]

```python
# Creating same df with different syntax and name for practice
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```
