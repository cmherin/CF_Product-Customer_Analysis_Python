2. If statements with user defined functions

```python
# Create subset - smaller dataset for processing 

df = df_ords_prod_merged[:1000000]
```

```python
df.shape
```

```python
# Create price labels for products - <$5 = 'low-range product', $5-$15 = 'mid-range product', >$15 'high-range product'

# define 

def price_label(row):
    if row['prices'] <= 5:
        return 'low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'mid-range product'
    elif row['prices'] > 15:
        return 'high-range product'
    else: return 'not enough data'
```

```python
# Add 'price_range' column to df

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check value counts on new price_range column

df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

:: This subset of data contains product prices within the condition of 'high-range product' in addition to 'not enough data'

4. If statements with for-loops

```python
# Create new 'df' whole (df_3) to use with for-loops

df_3 = df_ords_prod_merged
```

```python
# Dataframe shape

df_3.shape
```

```python
# Orders by day of the week

df_3['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create column showing what day (busiest, least busy, regularly busy) orders were placed 

result = []

for value in df_3['orders_day_of_week']:
    if value == 0:
        result.append('busiest day')
    elif value == 4:
        result.append('least busy')
    else:
        result.append('regularly busy')
```

```python
result
```

```python
# Add column to 'df_3'

df_3['busiest_day'] = result
```

```python
# Value count 'busiest_day'

df_3['busiest_day'].value_counts(dropna = False)
```

```python
# Compare columns 'busiest_day' to 'orders_day_of_week'

df_3['busiest_day'].shape
```

:: Columns 'busiest_day' and 'orders_day_of_week' match indicateing the for loop populated to each row containing a value for 'orders_day_of_week'

2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Create column for busy days

days_result = []

for value in df_3['orders_day_of_week']:
    if value == (0 or 1):
        days_result.append('busiest days')
    elif value == (4 or 3):
        days_result.append('slowest days')
    else:
        days_result.append('regularly busy')
```

```python
# Add column to df

df_3['busy_days'] = days_result
```

4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# How many orders happen on each hour of the day?

df_3['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Most orders = between 10 - 16

df_3.loc[(df_3['order_hour_of_day'] > 9) & (df_3['order_hour_of_day'] < 17), 'busiest_period_of_day'] = 'Most orders'
```

```python
# Average orders = more than 16

df_3.loc[(df_3['order_hour_of_day'] > 16), 'busiest_period_of_day'] = 'Average orders'
```

```python
# Average orders = between 7 - 9

df_3.loc[(df_3['order_hour_of_day'] <= 9) & (df_3['order_hour_of_day'] >=7), 'busiest_period_of_day'] = 'Average orders'
```

```python
# Fewest orders = less than 7

df_3.loc[(df_3['order_hour_of_day'] < 7), 'busiest_period_of_day'] = 'Fewest orders'
```

```python
# Busiest hours of the day stats

df_3['busiest_period_of_day'].value_counts(dropna = False)
```

:: The busiest period of time for orders is between 10am and 4pm
